a. Используя линейную регрессию, обучите модель с экспоненциальным ростом числа заболевших: y ~ exp(линейная функция от x), где x — номер текущего дня.

```python
class ExponRidge():
    def __init__(self, alpha=1, eps=0):
        self.alpha = alpha
        self.eps = eps
        
    def fit(self, X, y):
        y = np.log(np.array(y) + self.eps)
        X = np.array(X)
        ones = np.ones((X.shape[0], X.shape[1] + 1))
        ones[:, 1:] = X
        X = ones
        self.w = np.linalg.inv(X.T @ X + self.alpha*np.eye(2)) @ X.T @ y
        return self
        
    def predict(self, X):
        X = np.array(X)
        ones = np.ones((X.shape[0], X.shape[1] + 1))
        ones[:, 1:] = X
        X = ones
        return np.exp(X @ self.w) - self.eps
    
    def estimate_variance(self, X, y):
        return (np.log(self.predict(X)) - np.log(y)).std()**2
```

```python
def alpha_search(col_name, alphas, eps=0):
    best_alpha = alphas[0]
    best_error = np.inf
    for alpha in alphas:
        model = ExponRidge(alpha, eps=eps)
        model.fit(pd.DataFrame(df_train['day_number']), df_train[col_name])
        pred = model.predict(pd.DataFrame(df_train['day_number']))
        error = np.sqrt(((pred - df_train[col_name])**2).mean())
        print('Alpha: {:5}; Train error: {}'.format(alpha, error))
        if error < best_error:
            best_alpha = alpha
            best_error = error
    print('Best alpha: {}'.format(best_alpha))
```

```python
alphas = [0, 1, 10, 100, 1000, 2000, 4000, 5000, 6000, 7000, 10000, 20000, 50000]
alpha_search('new_cases', alphas, eps=1)
```

b. Найдите апостериорное распределение параметров этой модели для достаточно широкого априорного распределения. Требующееся для этого значение дисперсии шума в данных оцените, исходя из вашей же максимальной апостериорной модели (это фактически первый шаг эмпирического Байеса).

Как было показано на лекции, если априорного взять распределение $p(w) = \mathcal{N}(w|\mu_0, \Sigma_0)$, то апостериорное распределение имеет вид:
$$
p(w|D) = \mathcal{N}(w| \mu_N, \Sigma_N),
$$
где $\Sigma_N = (\Sigma_0^{-1} + \frac{1}{\sigma^2}X^T X)^{-1}$, $\mu_N = \Sigma_N(\Sigma_0^{-1}\mu_0 + \frac{1}{\sigma^2}X^T \ln{y})$, $\sigma^2$ - дисперсия шума в данных.  
Недостатком данной модели является предположение о независимости наблюдений, которое для данного временного ряда не выполняется.

В случае с гребневой регрессией мы используем априорное распределение $p(w) = \mathcal{N}(w|0, \sigma_0^2 I)$. Коэффициент регуляризации $\alpha = \frac{\sigma^2}{\sigma_0^2}$.  
Возьмём для примера модель для total_cases, и оценим для неё $\sigma^2$ основываясь на эмпирических данных.

```python
alpha = 4000
model = ExponRidge(alpha, eps=0)
model.fit(pd.DataFrame(df_train['day_number']), df_train['total_cases'])
var = model.estimate_variance(pd.DataFrame(df_train['day_number']), df_train['total_cases'])
var_zero = var / alpha
print(f'Estimated noise variance: {var}')
print(f'Estimated prior variance: {var_zero}')
```

Параметры апостериорного распределения:
$$
\Sigma_N = \left(\frac{1}{\sigma_0^2}I + \frac{1}{\sigma^2}X^T X\right)^{-1};
$$
$$
\mu_N = \Sigma_N\left(\frac{1}{\sigma^2}X^T \ln{y}\right);
$$

```python
X = np.ones((len(df_train), 2))
X[:, 1] = df_train['day_number']
sigma_posterior = np.linalg.inv(np.eye(2)/var_zero + (X.T @ X) / var)
mu_posterior = sigma_posterior @ (X.T @ np.log(df_train['total_cases'])) / var
print('Posterior mean:', mu_posterior)
print('Posterior covariance matrix:\n', sigma_posterior)
```

### Задание 3

Предсказания экспоненциальной модели наверняка получились грустными. Но это, конечно, чересчур пессимистично — экспоненциальный рост в природе никак не может продолжаться вечно. Кривая общего числа заболевших во время эпидемии в реальности имеет сигмоидальный вид: после начальной фазы экспоненциального роста неизбежно происходит насыщение. В качестве конкретной формы такой сигмоиды давайте возьмём форму функции распределения для гауссиана:
$$ \Phi(x) = \frac{1}{\sqrt{2 \pi}} \int_{-\infty}^{x}e^{-\frac{t^2}{2}}dt.$$
Естественно, в нашем случае сигмоида стремится не к единице, т.е. константа перед интегралом может быть произвольной (и её можно внести в экспоненту), а в экспоненте под интегралом может быть произвольная квадратичная функция от t.

a. Предложите способ обучать параметры такой сигмоидальной функции при помощи линейной регрессии.

Для того чтобы обучить параметры сигмоидальной функции можно продифференцировать исходный ряд: $y'_{(i)} = y_{(i)} - y_{(i-1)}$, а далее обучить модель $y' = у^{at^2 + bt + c}$, или что тоже самое $at^2 + bt + c = \ln{y'}$.

Для простоты используем регрессию с $L_2$ регуляризацией, то есть ограничим априорное распределение до $\mathcal{N}(w|0, \sigma_0^2 I)$. Это позволяет не знать дисперсию шума в данных заранее, а внести её в один гиперпараметр $\alpha = \frac{\sigma^2}{\sigma_0^2}$.

```python
class PhiRidge:
    def __init__(self, alpha=0):
        self.alpha = alpha
        
    def _differentiate(self, y):
        return y[1:] - y[:-1]
    
    def _make_x(self, X):
        X = np.array(X)
        res = np.ones((X.shape[0], 3))
        res[:, 1] = X
        res[:, 2] = X**2
        return res
    
    def _make_y(self, y):
        y = self._differentiate(np.array(y))
        y[y == 0] = 1
        return np.log(y)
        
    def fit(self, X, y):
        X = self._make_x(X)
        X = X[1:, :]
        y = self._make_y(y)
        self.X = X
        self.y = y
        self.w = np.linalg.inv(X.T @ X + self.alpha*np.eye(3)) @ X.T @ y
        return self
    
    def predict(self, X):
        X = self._make_x(X)
        y_pred = np.exp(X @ self.w)
        return y_pred.cumsum()
    
    def estimate_variance(self):
        errors = self.X @ self.w - self.y
        return errors.std()**2
```

b. Обучите эти параметры на датасете случаев коронавируса в России. Найдите апостериорное распределение параметров этой модели для достаточно широкого априорного распределения. Требующееся для этого значение дисперсии шума в данных оцените, исходя из вашей же максимальной апостериорной модели.

```python
def evaluate_model(alpha):
    model = PhiRidge(alpha)
    model.fit(df_train['day_number'], df_train['total_cases'])
    pred = model.predict(df['day_number'])
    plot_data('total_cases', pred=(df['day_number'], pred))
    
def alpha_search(col_name, alphas):
    best_alpha = alphas[0]
    best_error = np.inf
    for alpha in alphas:
        model = PhiRidge(alpha)
        model.fit(df_train['day_number'], df_train[col_name])
        pred = model.predict(df_train['day_number'])
        error = np.sqrt(((pred - df_train[col_name])**2).mean())
        print('Alpha: {:5}; Train error: {}'.format(alpha, error))
        if error < best_error:
            best_alpha = alpha
            best_error = error
    print('Best alpha: {}'.format(best_alpha))
```

```python
alphas = [0.001, 1, 2, 5, 10, 20]
alpha_search('total_cases', alphas)
```

Апостериорное распределение для весов будет аналогичным тому, что в предыдущем задании, за исключением того что $y$ меняется на $y'$:
$$
p(w|D) = \mathcal{N}(w| \mu_N, \Sigma_N),
$$
где $\Sigma_N = (\Sigma_0^{-1} + \frac{1}{\sigma^2}X^T X)^{-1}$, $\mu_N = \Sigma_N(\Sigma_0^{-1}\mu_0 + \frac{1}{\sigma^2}X^T \ln{y'})$, $\sigma^2$ - дисперсия шума в данных.  

```python
alpha = 0.001
model = PhiRidge(alpha)
model.fit(df_train['day_number'], df_train['total_cases'])
var = model.estimate_variance()
var_zero = var / alpha
print(f'Estimated noise variance: {var}')
print(f'Estimated prior variance: {var_zero}')
```

```python
X = model.X
y = model.y
sigma_posterior = np.linalg.inv(np.eye(3)/var_zero + (X.T @ X) / var)
mu_posterior = sigma_posterior @ (X.T @ y) / var
print('Posterior mean:', mu_posterior)
print('Posterior covariance matrix:\n', sigma_posterior)
```

Постройте эмпирическое предсказательное распределение, нарисуйте графики. Каков ваш прогноз числа случаев коронавируса в пессимистичном сценарии (90-й процентиль в выборке числа случаев)? В оптимистичном сценарии (10-й процентиль)?

```python
total_cases = []
num_trials = 10000
pred_length = 120
X = np.ones((pred_length, 3))
X[:, 1] = np.arange(1, pred_length+1)
X[:, 2] = np.arange(1, pred_length+1)**2
random_w = sps.multivariate_normal(mean=mu_posterior, cov=sigma_posterior).rvs(size=num_trials)
for w in random_w:
    y = np.cumsum(np.exp(X @ w))
    total_cases.append(y[-1])
    plt.plot(X[:, 1], y, 'k-', alpha=0.002, linewidth=0.5)
plt.plot(df_train['day_number'], df_train['total_cases'], label='train')
plt.plot(df_test['day_number'], df_test['total_cases'], label='test')
y = np.cumsum(np.exp(X @ random_w.mean(axis=0)))
plt.plot(X[:, 1], y, label='mean pred')
plt.xlabel('day number')
plt.ylabel('total cases')
plt.ylim((0, 600000))
plt.legend()
plt.show()
```
